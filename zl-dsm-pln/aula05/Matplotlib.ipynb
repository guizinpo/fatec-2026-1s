```python
faixa = np.arange(0, 12)
vendas = faixa ** 2
vendas
```

```python
despesas = vendas - 10
despesas[ despesas < 0 ] = 0
despesas[ 7 ] += 20
despesas
```

```python
plt.plot( meses, vendas )
plt.plot( meses, despesas )
plt.title("Vendas realizadas no ano")
plt.ylabel("Vendas")
plt.xlabel("Ano 2025")
plt.text(11, 120, "Redução do tarifaço")
plt.text(7, 60, "Treinamento dos funcionarios")
plt.legend()
# plt.savefig("vendas_2025.png")
# plt.show()
```
