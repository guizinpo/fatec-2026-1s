```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(binary=True, token_pattern=r"(?u)\b\w+\b")
```

```python
textos = ["O menino corre no parque.","Ele sorri enquanto brinca.","A mãe observa o menino de longe."]
```

```python
bow_binary = vectorizer.fit_transform( textos ) 
```

```python
vocabulario = vectorizer.get_feature_names_out()
vocabulario
```
