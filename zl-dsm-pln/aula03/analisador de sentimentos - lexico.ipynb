### Analisador de sentimento com Lemmatização

```python
texto = """O menino corre no parque. Ele sorri enquanto brinca. A mãe observa o menino de longe."""
```

```python
tabela = str.maketrans("", "", punctuation)
texto_limpo = texto.lower().translate(tabela)
```

```python
tokens = nlp(texto_limpo)
tokens
```

```python
soma = 0 
for token in tokens:
    palavra = token.lemma_
    if palavra in dicionario:
        valor = dicionario[palavra]
        print(token, "=>", palavra, ":", valor)
        soma += valor
print("O sentimento desta frase é: ", soma)
```
