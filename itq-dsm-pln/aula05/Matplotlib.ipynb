```python
np.random.seed( 50 )
valores_vendas = np.random.randint( 10, 200, (12, ), )
print("Valores de vendas: ", valores_vendas)

fator = np.random.random( (12, ) ) * 15
print("Fator: ",  fator)

gastos = valores_vendas - fator
print("Gastos: ", gastos)
```

```python
x = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'] 
```

```python
plt.plot(x, valores_vendas, marker="o", label="Vendas")
plt.plot(x, gastos, marker="*", label="Gastos")
# mudanças no pyplot (plt) 
# mudanças no pyplot (plt) 
# mudanças no pyplot (plt)
plt.ylabel("Vendas consolidadas")
plt.xlabel("Periodo de 2025")
plt.title("Vendas da filial de São Paulo")
plt.text(9, 5, "Redução de exportações\ndevido ao tarifaço")
plt.legend()
plt.show()
# plt.savefig("vendas-2025.png")
```

```python
faixa = np.arange(-100, 100, 10)
para_cima = faixa ** 2
para_baixo = -faixa ** 2
eixo_x = np.arange(0, 20)
```

```python
# plt.plot( para_cima, linestyle="dotted", color="blue" )
# plt.plot( para_baixo, linestyle="dashed", color="red", linewidth = 3 )
plt.scatter( eixo_x, para_cima, color="blue" )
plt.scatter( eixo_x, para_baixo, color="red" )
plt.axhline( 0 )
plt.axvline( 1000 )
plt.ylim(-2000, 2000)
plt.xlim(0, 20)
plt.show()
```

```python
plt.barh(x, valores_vendas, label="Vendas", color="blue")
plt.barh(x, gastos, label="Gastos", color="red")
plt.xlabel("Vendas consolidadas")
plt.ylabel("Periodo de 2025")
plt.title("Vendas da filial de São Paulo")
plt.legend()
plt.show()
```
