```python
from string import punctuation
```

```python
punctuation
```

```python
texto = """rato roeu a roupa do rei de Roma,
O rato roeu a roupa do rei da Rússia,
O rato roeu a roupa do RodovaIho...
O rato a roer roía
E a rosa Rita Ramalho do rato a roer se ria.
O rato roeu a roupa do rei de roma
a rainha com raiva roeu o resto."""
```

```python
texto = texto.lower()
texto = texto.replace("\n", " ")
```

```python
novo_texto = ""
for c in texto:
    if c not in punctuation:
        novo_texto += c
```

```python
texto
```

```python
novo_texto
```

```python
tokens = novo_texto.split(" ")  # ["rato", "roeu", "a", "roupa"]
```

```python
tokens
```

```python
numero_palavras_texto = len( tokens )
numero_palavras_texto
```

```python
vocabulario = []
for token in tokens:
    if token not in vocabulario:
        vocabulario.append( token )
vocabulario
```

```python
numero_palavras_vocabulario = len(vocabulario)
numero_palavras_vocabulario
```

```python
riqueza_lexical = numero_palavras_vocabulario / numero_palavras_texto
riqueza_lexical
```
