```python
import spacy
```

```python
nlp = spacy.load("pt_core_news_sm")
```

```python
texto = """O menino corre no parque. Ele sorri enquanto ele brinca. A mãe observa o menino de longe."""
```

```python
doc = nlp( texto )
```

```python
texto_lemma = []
for token in doc:
    # print(token, " - ", token.lemma_)
    texto_lemma.append( token.lemma_ )
" ".join(texto_lemma)
```
