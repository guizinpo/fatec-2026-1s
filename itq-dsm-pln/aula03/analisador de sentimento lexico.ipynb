```python
from string import punctuation
```

```python
import spacy
```

```python
nlp = spacy.load("pt_core_news_sm")
```

```python
lex = open( "C:\\git\\dados\\nlp\\lexico_v3.0.txt", "r", encoding="utf-8")
```

```python
palavras_valores = dict()
```

```python
lido = "\\"
linhas = 0
while lido != "":
# and linhas < 30:
    lido = lex.readline()
    lido = lido.replace("\n", "")
    colunas = lido.split(",")
    linhas += 1
    if len(colunas) > 2:
        # print(colunas[0], "=>", colunas[2])
        palavras_valores[colunas[0]] = colunas[2]
len()
```

```python
texto = """O menino corre no parque. Ele sorri enquanto ele brinca. A mãe observa o menino de longe."""
```

```python
trans = str.maketrans("", "", punctuation)
tokens = texto.lower().translate(trans).split(" ")
tokens
```

```python
# analisador de texto
soma = 0

for token in tokens:
    if token in palavras_valores:
        valor = int(palavras_valores[token])
        soma = soma + valor
        print(token, "=>", valor)

print("Analise do sentimento do texto ==> ", soma)
```

```python
tokens = nlp( texto )
```

```python
# analisador de texto lematizado
soma = 0

for token in tokens:
    lemma = token.lemma_
    if lemma in palavras_valores:
        valor = int(palavras_valores[lemma])
        soma = soma + valor
        print(lemma, "=>", valor)

print("Analise do sentimento do texto ==> ", soma)
```
