```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
textos = ["O menino corre no parque.",
"Ele sorri enquanto ele brinca.",
"A mãe observa o menino de longe."] 
```

```python
vectorizer = CountVectorizer( binary=False, token_pattern=r"(?u)\b\w+\b" )
```

```python
bow_binary = vectorizer.fit_transform( textos )
```

```python
bow_sparse = bow_binary.toarray()
bow_sparse
```
